```python
# ,play(),
# ,
# ?FadeOut()
#  ?,.
l = ["dracula", "fruity", "github-dark", "gruvbox-dark", 'inkpot', 'lightbulb', 'manni', 'material', 'monokai', 'murphy', 'native', 'one-dark']
l.append('paraiso-dark')
l.append('solarized-dark')
l.append('solarized-light')
l.append('stata-dark')
l.append('zenburn')
class Test(Scene):
    def construct(self):
        for style in l:
            my_code = Code(r'.\MyCode.py', background='window', formatter_style=f"{style}")
            text = Tex(f":{style}", tex_template = TexTemplateLibrary.ctex)
            VGroup(my_code, text).arrange(DOWN,buff = 0.3)
            self.play(
                Create(my_code),
                Create(text),
                run_time = 1.5
            )
            self.wait(2)
            self.play(
                FadeOut(my_code),
                FadeOut(text), 
                run_time = 1.5
            )
%manim $_RF Test
```
